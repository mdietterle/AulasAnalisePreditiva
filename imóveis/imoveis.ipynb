## Precisamos de números...

Alguns tipos de dados são categóricos, porém são interessantes para a análise dos dados.

Entretanto, não podemos usar dados categóricos nos nossos modelos, eles devem ser numéricos.

Desta forma, precisamos converter algumas características dos dados para números usando algumas técnicas.

O importante aqui, é que para o modelo não interessa (por exemplo) se a pessoa mora na zona urbana ou zona rural. No modelo matemático, podemos converter todos os registros de zona urbana para 0 e zona rural para 1. O modelo irá usar este dado para suas análises

Vamos ver a seguir, algumas técnicas de conversão de dados...

### Label encoder

No `LabelEncoder()`, as classes serão substituídas por números (ele atribuirá um valor 0 para a primeira classe encontrada, e toda vez que encontrar a mesma classe, irá atribuir a mesma classe. Fará o mesmo para a segunda, terceira, ... enésima classe que encontrar.

O Label encoder encontra-se no sklearn.

````python
label_encoder = LabelEncoder()
labels_zona = label_encoder.fit_transform(df.MSZoning)
labels_zona
````

````python
df['Zona'] = labels_zona
df.head(20)
````

**ATENÇÃO:** Precisamos tomar cuidado com esta abordagem, pois o nosso modelo pode assumir que um imóvel (neste nosso caso) que esteja na zona 4 é melhor que um que está na zona 2, simplesmente por ser um número maior. Também não é uma técnica que funcione com dados ordinais (por exemplo a posição de chegada de um piloto em uma série de corridas: Se na 1a. corrida ele chegou em 5o. lugar, esta vai ser a categoria 0 (a função não se importa com o dado em si, ela vai atribuir 0 para a primeira categoria que encontrar).

Quando temos mais de uma categoria, precisamos analisar com cuidado se este não é um caso a se prestar bastante atenção.

A conversão usando o `LabelEncoder()` funciona melhor quando temos somente duas categorias (ex: Rua pavimentada, Rua Não Pavimentada). 

Neste caso, para que o algoritmo não se atrapalhe (0 ainda é menor que 1), precisamos de mais um passo. Além de converter de uma variável categórica para numérica, precisamos mais um passo: Dividir esta informação em duas colunas, onde em uma assinalamos com 0 para não pavimentada e 1 para pavimentada, e na outra ao contrário, 1 para não pavimentada e 0 para pavimentada. Desta forma, equilibramos o algoritmo. Se tivermos mais de uma categoria na coluna original, serão criadas "n" categorias, uma para cada variação.

Para isto, aplicamos o método `OneHotEncoder()` na coluna resultant do método `LabelEncoder()`.

Novamente, o método `OneHotEncoder()` está no sklearn...

Vamos ao código:

````python
one_hot_encoder = OneHotEncoder()
feature_array = one_hot_encoder.fit_transform(df[['Zona']]).toarray()
feature_array
````

Agora que temos as categorias, vamos recolocar elas no nosso dataframe original:

````python
feature_labels = list(label_encoder.classes_)
features = pd.DataFrame(feature_array, columns=feature_labels)
features
````

Neste ponto, temos um Dataframe com as colunas como precisamos, agora, iremos concatenar com o Dataframe original:

````python
pd.concat([df, features], axis=1)
````

Como sempre, o professor ensina primeiramente uma técnica mais difícil e em seguida uma mais fácil (mas nem sempre passível de utilização/aplicação).

## Pandas `get_dummies()`

O método `get_dummies()` não necessita da aplicação do `label_encoder()` antes de aplicar o `get_dummies()`. Podemos aplicar o método diretamente na coluna com as strings que se deseja converter:

````python
dummy_features = pd.get_dummies(df['Street'])
pd.concat([df, dummy_features], axis=1)
````

Em alguns casos (gênero masculino e feminino, por exemplo), esta abordagem não funciona muito bem, porque se o registro for do gênero masculino, implica em não ser do gênero feminino (e vice versa). 

Neste caso, podemos "esquecer" uma das colunas gerada (a primeira), porque tendo somente uma já temos o resultado correto. Se desejarmos "esquecer" a primeira coluna, passamos o parâmetro `drop_first = True` para o método.

Porém precisamos tomar cuidado e analisar se realmente é o caso de esquecer ou não. (geralmente é indicado com dados altamente correlacionados).
